```python
"""
Day 1: DNA Sequence Analyzer
Author: Aqsa Hamdani  
Date: 2026-05-05
"""

def analyze_dna(sequence):
    """
    Analyze a DNA sequence and print statistics.
    """
    # Convert to uppercase for consistency
    sequence = sequence.upper()
    
    # Calculate basic statistics
    length = len(sequence)
    a_count = sequence.count('A')
    t_count = sequence.count('T')
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    
    # Calculate GC content
    gc_content = ((g_count + c_count) / length) * 100
    
    # Print results
    print("=" * 50)
    print("DNA SEQUENCE ANALYSIS")
    print("=" * 50)
    print(f"Sequence: {sequence}")
    print(f"Length: {length} base pairs")
    print(f"\nNucleotide Counts:")
    print(f"  A (Adenine):  {a_count}")
    print(f"  T (Thymine):  {t_count}")
    print(f"  G (Guanine):  {g_count}")
    print(f"  C (Cytosine): {c_count}")
    print(f"\nGC Content: {gc_content:.2f}%")
    print("=" * 50)

```

```python
# Test it!
test_sequence = "ATGCGATCGATCGATCG"
analyze_dna(test_sequence)
```

```python
# Try your own sequence
my_sequence = "ATGGCCATTGTAATGGGCCGC"
analyze_dna(my_sequence)
```
