Codon Frequency Analyzer

Build a tool that analyzes codon usage in a DNA sequence.

Requirements:

Take a DNA sequence (hardcode it, e.g. "ATGGCCTAAATGCCCATGTTTAAAGGG")

Split it into codons (groups of 3 bases)

Count how many times each unique codon appears

Print a formatted report

```python

# given dna sequence
dna = "ATGGCCTAAATGCCCATGTTTAAAGGG"
# for loop to split the dna into codons
for i in range(0, len(dna), 3):
    codon = dna[i:i+3]
    
# count how many times each unique codon appears
# creating a new dictionary to store the codon counts
codon_counts = {}
for i in range(0, len(dna), 3):
    codon = dna[i:i+3]
    if codon in codon_counts:
        codon_counts[codon] += 1
    else:
        codon_counts[codon] = 1 

print("CODON FREQUENCY ANALYSIS")
print("-" * 40)
print("Sequence:",dna)
print("Total Codons:", len(dna) // 3)
print("\n Codon Frequencies:")
# print each codon and its count
for codon in codon_counts:
    print(codon, ":", codon_counts[codon])
    
```
