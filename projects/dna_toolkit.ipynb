This is a real one. Build a small "DNA toolkit" — a notebook with 5 reusable functions that work together.

Required functions:

gc_content(sequence) — returns GC% (from Step 3)

is_valid_dna(sequence) — returns True/False (from Step 3)

reverse_complement(sequence) — returns reverse complement (from Step 3)

count_codons(sequence) — returns codon frequency dictionary (from Step 3)

analyze_sequence(sequence) — a master function that uses all the others and prints a full report

analyze_sequence("ATGGCCATGTAA")

```python
def gc_content(sequence):
    """Calculate GC content percentage."""
    sequence = sequence.upper()
    gc = sequence.count('G') + sequence.count('C')
    return (gc / len(sequence)) * 100
```

```python
def is_valid_dna(sequence):
    """Check if sequence contains only A, T, G, C."""
    sequence = sequence.upper()
    valid_bases = "ATGC"
    for base in sequence:
        if base not in valid_bases:
            return False
    return True
```

```python
def reverse_complement(sequence):
    """Return reverse complement of DNA."""
    sequence = sequence.upper()
    complement = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    rev_comp = ""
    for base in sequence[::-1]:  # reverse the sequence
        rev_comp = rev_comp + complement[base]
    return rev_comp
```

```python
def count_codons(sequence):
    """Count frequency of each codon."""
    sequence = sequence.upper()
    codon_counts = {}
    for i in range(0, len(sequence), 3):
        codon = sequence[i:i+3]
        if len(codon) == 3:  # only count full codons
            if codon in codon_counts:
                codon_counts[codon] += 1
            else:
                codon_counts[codon] = 1
    return codon_counts
```

```python
def analyze_sequence(sequence):
    """Complete sequence analysis using all other functions."""
    print("=" * 50)
    print("DNA SEQUENCE ANALYSIS REPORT")
    print("=" * 50)
    
    # Check validity first
    if not is_valid_dna(sequence):
        print("✗ Invalid DNA sequence!")
        return
    
    sequence = sequence.upper()
    
    # Basic info
    print(f"\nSequence: {sequence}")
    print(f"Length: {len(sequence)} bp")
    print(f"Valid DNA: ✓")
    
    # GC content
    gc = gc_content(sequence)
    print(f"\nGC Content: {gc:.2f}%")
    
    # Reverse complement
    rc = reverse_complement(sequence)
    print(f"Reverse Complement: {rc}")
    
    # Codon analysis
    codons = count_codons(sequence)
    print(f"\nCodon Frequencies:")
    for codon in codons:
        print(f"  {codon}: {codons[codon]}")
    
    print("=" * 50)


print(analyze_sequence("ATGGCCATGTAA"))
```
