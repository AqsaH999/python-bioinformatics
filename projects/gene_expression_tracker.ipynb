Gene Expression Tracker


```python
# A list of genes
genes = ["BRCA1", "TP53", "EGFR", "MYC", "KRAS"]
# A list of expression values
expression_values = [12.4 ,88.2 ,45.6 ,102.1 ,7.8]
#Print output 
print("Gene Expression Tracker")
print("=" * 30 )
# print gene and its expression value 
for gene, expression_value in zip(genes,expression_values):
    print(f"{gene} : {expression_value}")
# Total genes
print("Total genes",len(genes))
#Highest expression
print("Highest Expression:",max(expression_values))
#Lowest expression
print("Lowest expression:",min(expression_values))
#Average Expression
print("Average Expression:",sum(expression_values)/len(expression_values))
```
