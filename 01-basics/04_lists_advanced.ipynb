Concept: Lists in Loops 



Yesterday I used a loop to print items from a list. 


Today I'll learn to buildlists using loops 


— one of the most common patterns in bioinformatics.

Pattern 1. Loop through a list and do something with each item

```python
# list of genes 
genes = ["BRCA1", "TP53", "EGFR"]
# applying for loop
for gene in genes:
    print(f"Gene: {gene}")
```

Pattern 2. Create new list from existing list

```python
#pattern is found everywhere in bioinformatics
# exisiting list of genes
genes = ["brac1","tp53","egfr"]                     
# new list
uppercase_genes = []
# apply for loop
for gene in genes:
    (uppercase_genes.append(gene.upper()))
print(uppercase_genes)
```

Pattern 3. Nested lists (lists inside lists)

```python
 # A list of lists — each inner list is a gene with its info
genes_data = [
    ["BRCA1", "chr 17","cancer"],
    ["TP53", "chr17", "cancer"],
    ["EGFR", "chr17", "cancer"]
]
# Access the first gene's data
print(genes_data[0])
print(genes_data[0][0])
print(genes_data[0][1])
print(genes_data[0][2])
# Access the second gene's data
print(genes_data[1])
print(genes_data[1][0])
print(genes_data[1][1])
print(genes_data[1][2])
```

Basic Python Examples:

```python
# Pattern 1: Loop through a list
numbers = [10, 20, 30, 40]
for number in numbers:
    print(number * 2)
```

```python
# Pattern 2: Build a new list
numbers = [10, 20, 30, 40]
doubled = []
for num in numbers:
    doubled.append(num *2)
print("Doubled Numbers:",doubled)    
```

```python
# Pattern 3: Nested lists (a simple table)
students = [
    ["Alice", 85, 90],
    ["Bob", 78, 88],
    ["Charlie", 92, 95]
]
for student in students:
    name = student[0]
    score1 = student[1]
    score2 = student[2]
    avg_score = (score1+score2)/2   # don't use sum() , as it will sum the int and give an error
    print(f"Name: {name},  Average Score:{avg_score}")
```

Bioinformatics Application Examples: 

Example 1: Extract all sequences that start with start codon

```python
# the list of dna sequences
sequences = ["ATGGCCTAA", "ATGAAATGA", "GCGATGGCC", "ATGTTTAAA"]
# the new list
starts_with_atg =[]
# apply for loop
for seq in sequences:
    if seq.startswith("ATG"):
     starts_with_atg.append(seq)
print(starts_with_atg)





```

Example 2: Calculate GC content for multiple sequences

```python
# list of dna sequences
sequences = ["ATGGCC", "GCGCGC", "ATATAT"]
# the new list
gc_content =[]
# apply for loop
for seq in sequences:
    g_content = seq.count("G")
    c_content = seq.count("C")
    gc_percentage = (g_content + c_content)/len(seq)* 100
    gc_content.append(gc_percentage)

# Print each sequence with its GC content  
for seq, gc in zip (sequences, gc_content):
    print(f"{seq}: {gc:.1f}% GC")
   
```

Example 3: Nested list — gene database

```python
# Each gene is [name, chromosome, category]
gene_database =[
    ["BRCA1", "chr17", "cancer"],
    ["TP53", "chr17", "cancer"],
    ["INS", "chr11", "metabolic"],
    ["EGFR", "chr7", "cancer"],
    ["HBB", "chr11", "blood"]
]
# Find all cancer genes 
cancer_genes=[]
for gene in gene_database:
     name = gene[0]
     category = gene[2]
     if category == "cancer":
          cancer_genes.append(name)
print("Cancer Genes:", cancer_genes)

# Count how many genes are on chr17
chr17_count = 0                       #start counter on chr17
for gene in gene_database:
     chromosome = gene[1]
     if chromosome == "chr17":
      chr17_count +=1               # increment counter for each gene on chr17
print("Genes on chr17:" , chr17_count)
```

Practice Problem 3:

In the given gene database :

*Print the name of the longest gene

*Count how many cancer genes there are

*Calculate the average length of all genes

```python
# given gene database
genes = [
       ["BRCA1", 5711, "cancer"],
       ["TP53", 1182, "cancer"],
       ["INS", 333, "metabolic"],
       ["EGFR", 3633, "cancer"]
   ]
# a new list
all_lengths = []
# apply for loop to add all lengths
for gene in genes:
    gene_length = gene[1]
    all_lengths.append(gene[1])
# now find the max length
longest = max(all_lengths)
# which gene has that length
for gene in genes:
    if gene[1] == longest:
        print("Longest Gene:", gene[0], "\nLength:", longest)
#count thow many cancer genes are their
cancer_genes = 0
for gene in genes:
    disease = gene[2]
    if disease == "cancer":
        cancer_genes += 1
print("Number of Cancer Genes:", cancer_genes)  
# find the average length of all genes
average_length = sum(all_lengths) / len(all_lengths)
print("Average Gene Length:", average_length)
```

Practice Problem 4:

Given dna = "ATGGCCTAAATGCCCATGTTT", split it into codons (groups of 3) and 

store them in a list. Result should be: ["ATG", "GCC", "TAA", "ATG", "CCC", "ATG", "TTT"]


Hint: use a loop with range(0, len(dna), 3) to jump by 3s, then slice dna [ i:i+3]

```python
# dna sequence
dna = "ATGGCCTAAATGCCCATGTTT"
# a new list
grouped_list = []
# splitting dna into group of 3 codon's
for i in range(0, len(dna), 3):
    codon = dna[i:i+3]
    grouped_list.append(codon)
print(grouped_list)
```

Practice Problem 5:

(Harder) Given two nested lists:

gene_names = [["BRCA1", "TP53"], ["EGFR", "MYC"]]

Flatten this into a single list: ["BRCA1", "TP53", "EGFR", "MYC"]

You'll need a loop inside a loop (nested loops)

```python
# given nested list
gene_names = [["BRCA1", "TP53"], ["EGFR","MYC"]]
# a new list
flattened_genes = []
# apply for loop
for sublist in gene_names:
    for gene in sublist:
        flattened_genes.append(gene)
print(flattened_genes)
```
