A function is a reusable block of code that does one specific job.

 You write it once, then "call" it whenever you need it.

You've already been using functions all week: print(), len(), range(), sum(), .count() — 

these are all functions Python provides.

Why functions matter (especially in bioinformatics):

 Calculate GC content of many sequences. You'll parse many FASTA files. 
 
 You'll translate many DNA → protein. Writing the same code over and over 
 
 is error-prone and ugly. A function lets you write the logic once and reuse it.

# Examples

```python
#Example 1: Function with no input, no return
def say_hello():
    print("Hi there!")
say_hello()
```

```python
# Example 2: Function with input, no return
def square_and_print(n):
    print(n*n)
square_and_print(2)
square_and_print(5)
square_and_print(6)
```

```python
# Example 3: Function with input AND return
def square(n):
    return n*n
result = square(5)            #with return results hols the value , used in real world functions
print(result)

print(square(10))                #shortcut calling and printing fnc in one step 
print(square(20))

```

```python
# Example 4: Multiple parameters
def add(a,b):
    return a+b
result = add(25,55)
print(result)

print(add(100,67))
```

```python
#Example 5: Function calling another function
def double(n):
    return n*2

def quadruple(n):
    return double(double(n))          #function uses another function

print(quadruple(5))
print(quadruple(10))
print(quadruple(20))
```

```python
#Difference between print and return (beginers struggle with this!)

def with_print(n):
    print(n*2)

def with_return(n):
    return n*2

a = with_print(5)               #prints on screen but doesn't store
b = with_return(10)             #stores the value 

print(a)
print(b)

```

Bioinformatics Application :

```python
#Function 1: Calculate GC content
def gc_content(sequence):
    g_count = sequence.count("G")
    c_count = sequence.count("C")
    gc_percent = (g_count + c_count /len (sequence) * 100)
    return gc_percent

#Test it!
print(gc_content("ATGGCC")) 
print(gc_content("AAATTT"))
print(gc_content("GCGCGC")) 
```

```python
#Function 2: Check if a sequence is a valid DNA string
def is_valid_dna(sequence):
    valid_bases = ("A", "T", "G", "C")
    for base in sequence:
        if base not in valid_bases:
            return False
    return True             # Return can be used in the middle of a function, not just at the end. 
                            # As soon as Python hits return, it exits the function immediately.

#Test it!
print(is_valid_dna("ATGGCC")) 
print(is_valid_dna("ATGXCC"))
print(is_valid_dna("ATGNCC"))
```

```python
#Function 3: Reverse complement of a DNA sequence
def reverse_complement(sequence):
    complement = {"A": "T", "T": "A","G": "C","C":"G"}
    rev_complement = ""
    for base in sequence [::-1]:                          #It is slicing syntax for "reverse this string"
        rev_complement =  rev_complement + complement[base]
    return rev_complement
#Test it!
print(reverse_complement("ATGGCC"))
print(reverse_complement("AAATTT"))            #Panlindromic sequence
```

```python
#Function 4: Count codons in a sequence

def codon_count(sequence):
    codons = {}
    for i in range(0,len(sequence), 3):
        codon = sequence[i:i+3]
        if len(codon) == 3:                    # skip incomplete codons at end
            if codon in codons:
                codons[codon] += 1
            else: 
                codons[codon]= 1
    return codons   
#Test it!
print(codon_count("ATGGCCATGTAA"))        #odon freq analyzer now  packaged in a reusable fnc 
```

```python
# Write a function add_one(n) that takes a number and returns that number plus 1.
def add_one(n):
    return n + 1 

print(add_one(5))
print(add_one(10))     
```

```python
# Write a function count_a(sequence) that takes a DNA sequence and returns how many "A"s it contains.
def count_a(sequence):
    a_count = sequence.count("A")
    return a_count

print(count_a("ATGGCC"))
print(count_a("AAATTT"))
```

```python
# Write a function seq_length(sequence) that takes a sequence and returns its length. 
def seq_length (sequence):
    return len (sequence)

print(seq_length("ATGGCC"))
print(seq_length("AAATTT"))
```

```python
# Write a function is_long(sequence, threshold) that returns True if the sequence is longer than threshold and False otherwise.

# Example: is_long("ATGGCC", 5) → True
# Example: is_long("ATG", 5) → False
def is_long(sequence, threshold):
    if len(sequence) >= threshold:
        return True 
    else:
        return False
    
print(is_long("ATGGCC", 5))
print(is_long("ATG", 5))
```

```python
# Write a function at_content(sequence) that returns the AT content (the percentage of A's and T's). 
def at_content(sequence):
    a_count = sequence.count("A")
    t_count = sequence.count("T")
    at_percent = ( a_count + t_count ) / len(sequence) * 100
    return at_percent

print(at_content("ATGGCC"))
print(at_content("AAATTT"))

```

```python
# Write a function transcribe(dna) that takes a DNA sequence and returns the mRNA equivalent (replace every "T" with "U").

# Example: transcribe("ATGGCC") → "AUGGCC"

def transcribe(dna):
    mrna = dna.replace ("T", "U")
    return mrna

print(transcribe("ATGGCC"))
```
