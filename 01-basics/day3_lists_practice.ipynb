Examples for lists

```python
# A list of numbers
numbers = [1, 2, 3, 4, 5, 6]
#first element
print(numbers[0])
#last element
print(numbers[-1])
#length
print(len(numbers))
#sum
print(sum(numbers))
#max element
print(max(numbers))
```

```python
# A list of strings
names  = ["apple","banana", "cherry", "mango"]
#add an element 
names.append("orange")
print(names)
#remove an element
names.remove("cherry")
print(names)
```

Bioinformatics 
Example 1 :

```python
#A list of cancer genes
cancer_genes = ["BRCA1","TP53","EGFR", "MYC","KRAS"]
#Length of the list
print(len(cancer_genes))
#Find an element
print("EGFR" in cancer_genes)
#Add an element
cancer_genes.append("PTEN")
print(cancer_genes)
#Add a list of expression values
expression_values = [12.4, 88.2, 45.6, 102.1, 7.8]
print(expression_values)
# Print the highest expression value 
print(max(expression_values))
#Print the gene with highest expression value 
max_exp_gene = expression_values.index(max(expression_values))
print(cancer_genes[max_exp_gene])
#Average expression
average_expression = sum(expression_values) / len(expression_values)
print(average_expression)
```

```python
# A DNA sequence can be stored as a string OR a list of characters
dna_string = "ATGGCCTAA"
# convert string to list of chars
dna_list= list(dna_string)
print(dna_list)
#count the G's in the list
print(dna_list.count("G"))
```
